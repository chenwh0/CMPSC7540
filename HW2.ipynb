# **Installs & Imports**

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Define notations**

```python
def cost_function_J(theta: float) -> float:
    return (2 * theta - 3) ** 2 # J(θ) = (2θ - 3)²
def gradient_J(theta: float) -> float:
    return 4 * (2 * theta - 3) # dJ(θ)/dθ = 4 (2θ - 3)
```

#**Gradient descent**

```python
def gradient_descent(learning_rate=0.1, iterations=100) -> float: # Descend for 100 steps
    theta = 0 # Random guess to start
    thetas = [] # past thetas
    j_values = [] # Corresponding theta values
    for _ in range(iterations):
        gradient = gradient_J(theta) # Calculate gradient of cost function, plugging in theta value
        theta = theta - learning_rate * gradient # Gradient update = θ - α * dJ(θ)/dθ
        thetas.append(theta)
        j_values.append(cost_function_J(theta))
    return theta, thetas, j_values
```

```python
min_theta, thetas, j_values = gradient_descent() # Run gradient descent
minimum_J = cost_function_J(min_theta)
```

## Plot results

```python
plt.figure(figsize=(10, 6))
plt.plot(thetas, j_values, label='J(θ) = (2θ - 3)²')
plt.xlabel('θ')
plt.ylabel('J(θ)')
plt.title('Gradient Descent finding minimum of J(θ)')
plt.scatter(min_theta, cost_function_J(min_theta), color='red', label=f'Minimum of J(θ) ={min_theta}')
plt.legend()
plt.show()
print(f"Minimum at ({min_theta}, {minimum_J})")
```
