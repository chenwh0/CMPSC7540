# **Installs & Imports**

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

# **PCA step-by-step exercise**

```python
# 1. Dataset generation
random.seed(1)
x1 = [random.randint(1, 100) for i in range(10)]
x2 = [random.randint(1, 100) for i in range(10)]

x1 = np.array(x1)
x2 = np.array(x2)

# Data visualization
plt.scatter(x1,x2, color="hotpink")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Original Data")
plt.show()

# 3. Mean Centering
x1_mean = np.mean(x1)
x2_mean = np.mean(x2)
recentered_x1 = x1 - x1_mean
recentered_x2 = x2 - x2_mean

# 4. Covariance Matrix calculation
covariance_matrix = np.cov(recentered_x1, recentered_x2) # find covariance

# 5. Eigenvector and Eigenvalue Computation
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix) # find eigenvalues and eigenvectors

# 6. Principal component selection
index_of_max_eigenvalues = np.argmax(eigenvalues)
eigenvector_of_max = eigenvectors[0]

# 7. Dimensionality Reduction
reduced_x = np.dot(np.array([x1, x2]).T, eigenvector_of_max)

# 8. Data Reconstruction
reconstructed_x = (reduced_x[:, np.newaxis] * eigenvector_of_max) + np.array([x1_mean, x2_mean]) # Reshape reduced_x to be a column vector
reconstructed_x1 = reconstructed_x[:,0]
reconstructed_x2 = reconstructed_x[:,1]

# 9. Visualization of Results
plt.scatter(x1,x2, color="hotpink")
plt.scatter(reconstructed_x1, reconstructed_x2, color="deepskyblue")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("PCA")
plt.legend(["Original Data", "Reconstructed Data"])
plt.show()
```
